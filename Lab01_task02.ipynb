```python
# Step 0: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')  # This will ask you to authorize access

file_path ='/content/drive/My Drive/Colab Notebooks/Au_nanoparticle_dataset.csv'

# Load into a pandas dataframe
df = pd.read_csv(file_path)

# Step 2: Create a new dataframe with only required columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Step 3: Display the first 20 samples
print("First 20 samples:")
print(new_df.head(20))

# Step 4: Calculate mean, std, and quartile values
stats = new_df.describe()  # describe() gives count, mean, std, min, 25%, 50%, 75%, max
print("\nStatistics for the 4 features:")
print(stats)

# Step 5: Plot histograms of each feature in 1x4 layout
plt.figure(figsize=(20, 5))
for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    sns.histplot(new_df[column], kde=False, bins=30, color='skyblue')
    plt.title(f'Histogram of {column}')
plt.tight_layout()
plt.show()



# Step 6: Pairplot visualization with seaborn
sns.pairplot(new_df)
plt.show()

# Step 7: Advanced PairGrid customization
g = sns.PairGrid(new_df)

# a. Diagonal: histogram + KDE
g.map_diag(sns.histplot, kde=True, color='lightgreen')

# b. Upper: histogram
g.map_upper(sns.histplot, color='lightblue')

# c. Lower: bivariate KDE
g.map_lower(sns.kdeplot, fill=True, cmap="Blues")

plt.show()
```
