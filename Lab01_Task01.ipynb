```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd

# Create a random 4x4 array from an exponential distribution
exp_array = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Distribution Array:\n", exp_array)


# Create a random 100000x1 array from exponential distribution
exp_data = np.random.exponential(scale=1.0, size=(100000, 1))
uniform_data = np.random.rand(100000, 1)  # Uniform
normal_data = np.random.randn(100000, 1)  # Normal

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(exp_data, density=True, bins=80, histtype="step", label="Exponential", color="red")
plt.hist(uniform_data, density=True, bins=80, histtype="step", label="Uniform", color="blue")
plt.hist(normal_data, density=True, bins=80, histtype="step", label="Normal", color="green")

plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# 3D plot for Z = X^2 + Y^2 with x,y in [-5,5]
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Z = X + Y")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()


```

```python
# Download the CSV
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset using pandas
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin1')
# Display first 5 rows
df.head()

# Select numeric Pokmon stats for correlation
cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stats_df = df[cols].dropna()

# Pearson correlation
pearson_corr = stats_df.corr(method="pearson")


# Spearman correlation
spearman_corr = stats_df.corr(method="spearman")
```

```python
# Pearson heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()

# Spearman heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()



```
